```python
spec = librosa.feature.melspectrogram(y[150000:500000],sr)
```

```python
spec2 = librosa.feature.melspectrogram(y2[150000:500000],sr2)
```

```python
spec3 = librosa.feature.melspectrogram(y3[150000:500000], sr3)
```

```python
spec.shape
```

```python
spec2.shape
```

```python
spec3.shape
```

```python
import librosa.display
```

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(spec,
                                              ref=np.max),
                          y_axis='mel', fmax=8000,
                          x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram')
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(spec2,
                                              ref=np.max),
                          y_axis='mel', fmax=8000,
                          x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram')
plt.tight_layout()
```

```python
plt.figure(figsize=(10, 4))
librosa.display.specshow(librosa.power_to_db(spec3,
                                              ref=np.max),
                          y_axis='mel', fmax=8000,
                          x_axis='time')
plt.colorbar(format='%+2.0f dB')
plt.title('Mel spectrogram')
plt.tight_layout()
```
